# Check out the Avacado dataset before making the Dash dashboard for it #

### Import packages and read in dataset ###

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('avocado.csv')
df.head()
```

```python
cities = df.geography.unique()
len(cities)
```

```python
df[df.geography=='Albany']
```

### Understanding the data ###

Some things about the data are:
- each place has average prices for the day listed once a week, one for conventional and one for organic
- there are 556 rows of data for each place, so there are 556/2 = 278 dates (weeks) of data
- the time spans from January, 2015 to May, 2020

### The dashboard ###

The dashboard should contain a graph of the average price over time of avocados. <br><br>
It should be a line graph, one for organic and one for conventional. <br>
Users should have a dropdown menu to pick the geographical region. <br><br>
Details of the graph:
- have a legend for avocado type
- title should say *Avocado Prices in -place-* where *-place-* is what the user selected
- date is along the *x-axis* and average price is along the *y-axis*

## Practice graph for Pittsburgh ##

```python
# subset data for Pittsburgh
pgh_df = df[df.geography == 'Pittsburgh']
city_title = 'Avocado Prices in Pittsburgh'

# graph Pittsburgh data
fig = px.line(pgh_df, x='date', y='average_price', color='type', title = city_title)
fig.show()
```
